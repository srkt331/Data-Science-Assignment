#Hypothesis Test Exercise 1

```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
```

```python
data=pd.read_csv('Cutlets.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
# paired two-sample t-test
stats.ttest_ind(data['Unit A'], data['Unit B'])
```

# from the above test we got p-value is 0.47 which is greater than alpha (0.05)
# hence H0 (Null Hypothesis) is True

#Hypothesis Test Excercise 2

```python
data1=pd.read_csv('LabTAT.csv')
```

```python
data1.head()
```

```python
data1.describe()
```

```python
stats.f_oneway(data1["Laboratory 1"],data1["Laboratory 2"],data1["Laboratory 3"],data1["Laboratory 4"])
```

# from the above test we got p-value is 2.11e-57 which is less than alpha (0.05)
# hence HA (Alternative Hypothesis) is True

#Hypothesis Test Excercise 3

```python
data2=pd.read_csv('BuyerRatio.csv')
```

```python
data2
```

```python
data2.describe()
```

```python
stats.chi2_contingency([data2["East"], data2["West"], data2["North"], data2["South"]])
```

# from the above test we got p-value is 0.66 which is greater than alpha (0.05)
# hence H0 (Null Hypothesis) is True
